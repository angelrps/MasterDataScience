```python
#RECAPITULACION
import pandas as pd
pd.set_option('display.max_columns', None)

bi = pd.read_csv('bookings.csv.bz2', sep='^', usecols=['arr_port', 'pax', 'year'], nrows=9999, chunksize=3000)

#data frame para guardar resultados de cada trocito
all_chunks=pd.DataFrame()

for i, b in enumerate(bi):
    print(i, len(b))    
    b=b.dropna()
    b=b[b['year']==2013]
    del b['year']
    result=b.groupby(['arr_port']).sum()
    
    all_chunks=all_chunks.append(result)
    
result_all = all_chunks.reset_index().groupby(['arr_port']).sum().sort_values(by='pax', ascending=False).reset_index()
result_all.head(10)
```

### Pre bonus solution

Añado el %%time, quito nrows y aumento el chunk size

ESTO ES EL RESULTADO DEL EJERCICIO 2

```python
%%time
import pandas as pd
pd.set_option('display.max_columns', None)

bi = pd.read_csv('bookings.csv.bz2', sep='^', usecols=['arr_port', 'pax', 'year'], chunksize=1000000)

#data frame para guardar resultados de cada trocito
all_chunks=pd.DataFrame()

for i, b in enumerate(bi):
    print(i, len(b))    
    b=b.dropna()
    b=b[b['year']==2013]
    del b['year']
    result=b.groupby(['arr_port']).sum()
    
    all_chunks=all_chunks.append(result)
    
result_all = all_chunks.reset_index().groupby(['arr_port']).sum().sort_values(by='pax', ascending=False).reset_index()
result_all.head(10)
```

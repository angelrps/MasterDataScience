````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

#### Exercise:
Given this code:
```python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.show()
```

Reproduce this plot:

![Exercise](./sine_exercise.png)

````python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
#fill first curve
plt.fill_between(X, Y+1, 1, alpha=0.3, color='blue')
#fill second curve red area
plt.fill_between(X, Y-1, -1, alpha=0.3, color='red', where=Y-1<-1)
#fill second curve blue area
plt.fill_between(X, Y-1, -1, alpha=0.3, color='blue', where=Y-1>-1)
#remove ticks
plt.xticks([])
plt.yticks([])
plt.savefig('sine_exercise.png')
````

#### Exercise:
Given this code:
```python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.show()
```

Reproduce this plot:

![Exercise](./sine_exercise.png)

````python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
#fill first curve
plt.fill_between(X, Y+1, 1, alpha=0.3, color='blue')
#fill second curve red area
plt.fill_between(X, Y-1, -1, alpha=0.3, color='red', where=Y-1<-1)
#fill second curve blue area
plt.fill_between(X, Y-1, -1, alpha=0.3, color='blue', where=Y-1>-1)
#remove ticks
plt.xticks([])
plt.yticks([])
plt.savefig('sine_exercise.png')
````

axes2=f.add_subplot(3,3,8) añade un axes dividiendo la figure en una rejilla de 3x3 y poniendo el axes en el recuadro posición 8

````python
xs = np.linspace(0, 10)
ys=xs**2

f = plt.figure(figsize=(10,8))

axes1=f.add_subplot(1,1,1)
axes2=f.add_subplot(3,3,8)
axes1.scatter(xs**2, ys)
````

f, ax = plt.subplots(3,1,sharex=True) crea una figura y la divide en 3 rows y 1 column, creando entonces 3 axes. Después podemos seleccionar los axes creados indexando.

````python
xs = np.linspace(0, 10)
ys=xs**2

f, ax = plt.subplots(3,1,sharex=True)
ax[0].scatter(xs**2,ys)
ax[2].scatter(xs,ys)
````
